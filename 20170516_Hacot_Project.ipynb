```python
corrmat = dfc.corr()
k = 34#number of variables for heatmap
cols = corrmat.nlargest(k, 'First sexual intercourse')['First sexual intercourse'].index



cm = dfc[cols].corr()

plt.figure(figsize=(9,9))

sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, cmap='hot' ,annot=True,vmin=0,vmax =1, square=True, annot_kws={'size': 5},
                 yticklabels = cols.values, xticklabels = cols.values)
#annot_kws={'size':} 
plt.show()
```
